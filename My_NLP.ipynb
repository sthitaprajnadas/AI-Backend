```python
import pandas as pd

df = pd.DataFrame({
    'mycol':['When life gives you lemons, make lemondaes!', 'She bought 2 lemons for $1 from maven market','A dozen lemons will make gallons of \xa0 lemondae!![All recipes]']
})
```

```python
import spacy
phrase = "i am seeling lemons for $5 today"
spacy = spacy.load('en_core_web_sm')
doc = spacy(phrase)
alltokens = [token.text for token in doc ]
print(alltokens)
tokens_nostop = [token.text for token in doc if not token.is_stop]
print(tokens_nostop)
tokens_lemmatized = [token.lemma_ for token in doc]
print(tokens_lemmatized)
```
